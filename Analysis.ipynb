# Combine Data

### Create Filtered Dataset

```python
# Remove all columns from fellowContact that also exist in fellowPitch, except for "PitchId" (used as the join key between datasets).
    # This prevents duplicate columns from appearing after merging the data.
matching = fellowContact.columns.intersection(fellowPitch.columns).difference(["PitchId"])
fellowContactClean = fellowContact.drop(columns = matching)
```

### Combine Contact and Pitch Datasets

```python
# Merge fellowPitch and fellowContactClean DataFrames on the "PitchId" column
    # Use a left join to keep all records from fellowPitch.
combinedPitchContact = pd.merge(fellowPitch, fellowContactClean,
    on = "PitchId",
    how = "left",
)
```

```python
# View top 5 rows of the dataset
combinedPitchContact.head()
```

```python
# View description of each column in dataset
combinedPitchContact.describe(include='all')
```

```python
# Check if the merged dataset has the same number of rows as fellowPitch dataset since a left join was performed.
print(f"Rows in merged dataset: {len(combinedPitchContact)}")
print(f"Rows in fellowPitch dataset: {len(fellowPitch)}")
```

```python
# Ensure there are no duplicate PitchIds
duplicateColumnsPitchIds = combinedPitchContact[combinedPitchContact.duplicated(subset="PitchId", keep=False)]
print(f"Duplicate PitchIds: {len(duplicateColumnsPitchIds)}")
```

### Export Combined Data

```python
# Export new merged dataset to csv file
    # Set index = False to prevent row index column to be exported to the csv file
    # Use na_rep = "NULL" to convert all NaN values in the dataset to NULL. If not done NaN values in csv are empty cells
combinedPitchContact.to_csv("3002223_CombinedData.csv", index = False, na_rep = "NULL")
```

# Calculate AVG Exit Velocity of Batted Ball

```python
# Select only the columsn needed to calculate AVG Exit Velo
exitVeloData = combinedPitchContact[["PitchId", "BatterTeam", "PitchCall", "ExitSpeed"]]

# Filter data to only include PitchCalls InPlay.
exitVeloData = exitVeloData[exitVeloData["PitchCall"] == "InPlay"]

# View exitVeloData
exitVeloData
```

```python
# GroupBy BatterTeam and calculate the mean of ExitSpeed.
    # SPR_CAR = Home Team
    # WIC_SUR = Away Team
avgExitVeloTeam = exitVeloData.groupby("BatterTeam")['ExitSpeed'].mean()

# Print results
print(avgExitVeloTeam)
```

```python
# RESULT
# Home Team AVG Exit Velo - 88.055246 mph
# Away Team AVG Exit Velo - 82.228852 mph
```
